```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
```

# Filtro FIR (Passa-Baixas) para 640 Hz

```python
# Carregar o arquivo de udio
fs, sinal = wavfile.read('sinal_1.wav')
sinal = sinal / np.max(np.abs(sinal))  # Normalizar o sinal

# Filtro FIR passa-baixas com frequncia de corte em 640 Hz
cutoff = 640 / (fs / 2)
numtaps = 80  # Tamanho do filtro
b = signal.firwin(numtaps, cutoff, window='hamming', pass_zero=True)
sinal_filtrado = signal.lfilter(b, 1, sinal)

# Vetor de tempo
tempo = np.arange(len(sinal)) / fs

# Resposta em frequncia do filtro
w, h = signal.freqz(b, worN=8000)

# Criar subplots lado a lado
fig, axs = plt.subplots(2, 2, figsize=(14, 6))

# Plot do sinal original no tempo (lado esquerdo)
axs[0, 0].plot(tempo, sinal)
axs[0, 0].set_xlim(0, 0.05)  # Zoom nos primeiros 10 ms
axs[0, 0].set_title('Sinal Original (Tempo)')
axs[0, 0].set_xlabel('Tempo [s]')
axs[0, 0].set_ylabel('Amplitude')
axs[0, 0].grid(True)

# Plot do sinal filtrado no tempo (lado esquerdo)
axs[1, 0].plot(tempo, sinal_filtrado, color='orange')
axs[1, 0].set_xlim(0, 0.05)  # Zoom nos primeiros 10 ms
axs[1, 0].set_title('Sinal Filtrado (640 Hz Passa-baixas)')
axs[1, 0].set_xlabel('Tempo [s]')
axs[1, 0].set_ylabel('Amplitude')
axs[1, 0].grid(True)

# Plot da resposta em frequncia (ganho) (lado direito)
axs[0, 1].plot(w * fs / (2 * np.pi), 20 * np.log10(abs(h)), color='blue')
axs[0, 1].set_title('Resposta do Filtro (640 Hz Passa-baixas) - Ganho')
axs[0, 1].set_xlabel('Frequncia [Hz]')
axs[0, 1].set_ylabel('Ganho [dB]')
axs[0, 1].grid(True)
axs[0, 1].set_xlim(0, 6000)
axs[0, 1].set_ylim(-100, 5)

# Plot da resposta em fase (lado direito)
axs[1, 1].plot(w * fs / (2 * np.pi), np.unwrap(np.angle(h)), color='green')
axs[1, 1].set_title('Resposta do Filtro (640 Hz Passa-baixas) - Fase')
axs[1, 1].set_xlabel('Frequncia [Hz]')
axs[1, 1].set_ylabel('Fase [radianos]')
axs[1, 1].grid(True)
axs[1, 1].set_xlim(0, 6000)

plt.tight_layout()
plt.show()
```
